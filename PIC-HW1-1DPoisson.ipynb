```python
import numpy as np
from matplotlib import pyplot as plt
from jupyterthemes import jtplot
jtplot.style(theme='monokai')
jtplot.style(ticks=True, grid=True, figsize=(10, 8))
import numba
```

```python
# Global vars (parameters & constants)
RHO = 1.00e+10
EPS_0 = 8.854187e-12
Q = 1.602176e-19

num_nodes = 51
dx = 0.02
ts_max = 5000  # number of time steps


def initArrays():
    #  Initialization
    x = np.arange(0, num_nodes) * dx
    phi = np.zeros(num_nodes)
    b = np.zeros(num_nodes)

    # Imposing boundary conditions
    phi[0] = 1
    phi[num_nodes - 1] = 1
    b[1:num_nodes - 1] = -Q * (RHO / EPS_0)
    b[0] = 0
    b[num_nodes - 1] = 0

    return x, phi, b


def main():
    """1D Poisson solver using finite differencing."""
    x, phi, b = initArrays()
    for step in range(ts_max):
        for i in range(1, num_nodes - 1):
            phi[i] = 0.5 * (phi[i - 1] - dx**2 * b[i] + phi[i + 1])
        if step % 100 == 0:
            plt.plot(x, phi)


if __name__ == "__main__":
    #  Set up plotting figure
    plt.figure(1, dpi=200)
    plt.ylim([0, 25])
    plt.xlabel('x')
    plt.ylabel(' phi ')

    main()
```
